La función también maneja errores de solicitud HTTP e imprime mensajes de error y continúa con el proceso para otras categorías si es posible.  
  Básicamente, se encarga de obtener datos de múltiples ítems para una lista de categorías específicada. Itera sobre cada categoría para obtener los datos de los ítems correspondientes y utiliza la función *get_description(item_id)* para obtener la descripción de cada ítem. Luego, almacena los datos de los ítems en un DataFrame de **Pandas** y los guarda en un archivo CSV llamado *'datos.csv'*.

```python
def obtener_descripcion(item_id):
    # Se construye la URL de la solicitud a la API de MercadoLibre para obtener la descripción de un item especifico.
    url = f'https://api.mercadolibre.com/items/{item_id}/description'
    # Se realiza una solicitud HTTP GET a la URL construida para obtener los datos de la descripcion del ítem.
    res = rq.get(url)
    # Se verifica si la solicitud fue exitosa (código de estado HTTP 200), lo que significa que se obtuvieron los datos de la descripción del item correctamente.
    if res.status_code == 200:
        # Si la solicitud fue exitosa, se convierten los datos de respuesta JSON a un diccionario de Python.
        data = res.json()
        #se extrae el texto plano de la descripcion del item.
        return data.get('plain_text', '')
    # En caso de que no se cumpla la condición se agrega otra condicion
    else:
        # En caso de error en la solicitud de obtener la descripción del producto, devuelve un texto indicando con que id del producto ocurrio el error
        print(f'Error al obtener la descripción del producto {item_id}')
        #En este caso, si la solicitud HTTP para obtener la descripcion del producto no fue exitosa (por ejemplo, debido a problemas de conexión o a que el producto no tiene una descripción), la función devuelve una cadena vacía ('').
        return ''

def items(categorias, limit):
    # Lista donde se almacenan los datos de todos los items obtenidos de la API
    all_items = []
    
    # For en caso de que se consulte mas de una categoria
    for cate_id in categorias:
        # Reiniciamos el offset para cada categoría
        offset = 0 
        # Iterarcion hasta lograr alcanzar el limite de items establecido en n_items o hasta que no haya mas resultados
        while True:
            # URL de la solicitud a la API
            url = f'https://api.mercadolibre.com/sites/MLA/search?category={cate_id}&limit={limit}&offset={offset}'
            # Se realiza la solicitud de HTTP a la API, para que el servidor los datos asociados y los almacene en res
            res = rq.get(url)
            # Se verifica si la solicitud fue exitosa ya que el codigo 200 en HTTP significa OK o Exito, es decir, que la solicitud fue procesada con exito
            if res.status_code == 200:
                # Convertimos la respuesta JSON a un diccionario de Python
                data = res.json()
                # Se obtiene los resultados de los items de la respuesta JSON, o una lista vacia si no hay resultados
                results = data.get('results', [])
                
                for item in results:
                    # Se obtiene el id del producto obtenido de la consulta en la categoria
                    item_id = item['id']
                    # Se obtiene la descripcion del producto
                    description = obtener_descripcion(item_id)
                    # Se agrega la descripcion al diccionario de item
                    item['description'] = description
                    # Se extrae el numero de items que contiene cada categoria
                    item['paging_total'] = data["paging"]["total"] 
                    item['paging_primary_results'] = data["paging"]["primary_results"]
                    item['paging_offset'] = data["paging"]["offset"]
                    item['paging_limit'] = data["paging"]["limit"]
                # Se agregan los datos de los multiples items a la lista
                all_items.extend(results)
                # Se actualiza el offset para la siguiente solicitud
                offset += limit
                # Se adiciona la condicion porque los n_item es el total de productos por categoria, por lo que no tendria por que salir mas productos
                if len(results) > limit:
                    # detener bucle
                    break
            # Nos arroja un mensaje de error si la solicitud no tuvo exito
            else:
                # Imprimir un texto si fue error indicando la extracción de que producto y que categoria quedo
                print(f'Error al obtener los item {item_id} de la categoría {cate_id}')
                # detener bucle
                break
   
    # Se crea un DataFrame de Pandas con todos los items obtenidos
    dt_items = pd.DataFrame(all_items)
    # Exportamos el DataFrame a un archivo CSV
    dt_items.to_csv('datos_all.csv', sep=';', decimal=',', index=False)

# {'id': 'MLA1039', 'name': 'Cámaras y Accesorios'},
# {'id': 'MLA1051', 'name': 'Celulares y Teléfonos'},
# {'id': 'MLA1648', 'name': 'Computación'},
# {'id': 'MLA1144', 'name': 'Consolas y Videojuegos'},
# {'id': 'MLA1000', 'name': 'Electrónica, Audio y Video'},

# Especificar las categorias
categorias = ['MLA1000','MLA1000','MLA1039', 'MLA1051', 'MLA1648','MLA1144','MLA1000']
limit = 50
items(categorias, limit)
```

```python
def extract_columns(dt_ini, variables2):
    # Realiza la iteracion del codigo de cada una de las variables especificadas
    for variable in variables2:
        # Se hace una lista vacia que se utiliza para almacenar los diccionarios extraidos de la cadena de texto de las columnas especificadas
        dic = []
        # Realiza la iteracion sobre cada variable especificada
        for d in dt_ini[variable]:
            # Verificar si las variables no son una cadena vacía ni un valor nulo
            if d != "" and pd.notnull(d):
                # Se inicia un bloque que se utiliza para manejar las excepciones
                try:
                    # Se verifica si la variable es una cadena de texto
                    if isinstance(d, str):
                        # Intenta analizar la cadena como un diccionario JSON
                        dic_ana = json.loads(d)
                        # Se verifica  si el diccionario analizado no es None, para evitar agregar valores None
                        if dic_ana is not None:
                            # Se agrega el diccionario a la lista
                            dic.append(dic_ana)
                    # En caso de que la variable no contenga una cadena de texto valida, simplemente agrega el elemento a la lista
                    else:
                        # Si el valor no es una cadena, simplemente agrégalo a la lista
                        dic.append(d)
                # Se capturan las excepciones que pueden ocurrir durante el proceso de carga del JSON
                except (json.JSONDecodeError, TypeError):
                    # Si ocurre un error al analizar la cadena, agrega un diccionario vacío a la lista
                    dic.append({})
            # Si el elemento es una cadena vacia o un valor nulo, agrega un diccionario vacio a la lisa
            else:
                # Se agregar un diccionario vacío para las cadenas vacías o valores nulos
                dic.append({})  
        # Se verifica si la lista dic contiene elementos
        if dic:  
            # Se crea un DataFrame utilizando los diccionarios almacenados en dic
            dat = pd.DataFrame(dic)
            # Se crea un diccionario de nombres teniendo como sufijo el nombre de la variable de la que se extrae la información y como prefijo el nombre de la variable extraida
            nombres = {col: f"{variable}_{col}" for col in dat.columns}
            # Renombrar las columnas del DataFrame con el formato despecificado anteriormente
            dat = dat.rename(columns=nombres)
            # Se agrega las columnas extraidas al DataFrame original
            dt_ini = pd.concat([dt_ini, dat], axis=1)
            # Se elimina la columna inicial de la que se extraen la informacion
            dt_ini = dt_ini.drop(variable, axis=1)
    # Devuelve el DataFrame modificado, depues de procesar todas las variables
    return dt_ini

# Obtener una lista de nombres de variables que comienzan con "attributes_"
variables2 = list(filter(lambda col: col.startswith('attributes_'), dt_ini.columns))
# Se llama a la funcion para procesar todo
dt_ini = extract_columns(dt_ini, variables2)
```

### Estadísticas descriptivas

Las estadísticas descriptivas son herramientas fundamentales en el análisis de datos que proporcionan resúmenes numéricos y visuales de las características fundamentales de un conjunto de datos. Estas estadísticas son especialmente útiles durante la etapa de entendimiento de datos para comprender la distribución, la dispersión y otras propiedades importantes de las variables que componen un conjunto de datos.

Las variables se pueden clasificar en dos categorías principales: variables categóricas y variables numéricas.

- **Variables categórica**: Son variables que representan categorías o grupos y tienen un conjunto finito y discreto de valores. Por ejemplo, el género (masculino/femenino), el estado civil (soltero/casado/divorciado), o el tipo de producto (A/B/C). Es importante determinar si una variable es categórica para aplicar métodos de análisis apropiados y comprender la distribución de las categorías, aunque en muchos casos suelen codificar una variable categórica (ejemplo: 1-Masculino y 2-Femenido, siendo genero 1 y 2), pero siguen siendo categóricas. Las estadísticas descriptivas más comunes incluyen:
  - **Número de categorías únicas**: Indica la cantidad de categorías diferentes presentes en la variable.
  - **Porcentaje de categorías**: Muestra la proporción de cada categoría con respecto al total de observaciones.
  - **Recuento de categorías más frecuentes**: Identifica las categorías más comunes y su frecuencia relativa en el conjunto de datos.


- **Variables numéricas**: Son variables que representan cantidades numéricas y pueden ser continuas o discretas. Por ejemplo, la edad, el precio de un producto, o la cantidad de productos vendidos. Es fundamental identificar si una variable es numérica para calcular estadísticas descriptivas relevantes y realizar análisis numéricos apropiados.Algunas de las estadísticas descriptivas comunes para variables numéricas incluyen:  
  - **Promedio**: Es la suma de todos los valores dividida por el número total de observaciones. Proporciona una medida de tendencia central.
  - **Mediana**: Es el valor que divide el conjunto de datos ordenados en dos partes iguales. Es menos sensible a valores extremos que la media.
  - **Mínimo y Máximo**: Son los valores más pequeño y más grande en el conjunto de datos, respectivamente.
  - **Desviación Estándar**: Es una medida de dispersión que indica la cantidad de variabilidad o dispersión de los datos con respecto a la media.
  - **Varianza**: Es el promedio de las diferencias al cuadrado entre cada valor y la media. Proporciona una medida de dispersión similar a la desviación estándar.

Es importante sacar estadísticas descriptivas durante la etapa de entendimiento de datos porque proporcionan una visión general de la distribución y la estructura de los datos. Estas estadísticas ayudan a identificar patrones, tendencias, valores atípicos y posibles problemas en los datos, lo que facilita la toma de decisiones informadas en análisis posteriores, como la modelización predictiva, la segmentación de datos y la generación de insights significativos para la toma de decisiones empresariales.

#### Nulos, vacíos y atípicos

El análisis de los nulos, vacíos y valores atípicos es esencial en la fase de entendimiento de datos debido a su impacto significativo en la calidad y la integridad de los datos. 

- **Nulos y vacíos**: Representan la ausencia de datos en un conjunto de datos. Pueden surgir debido a errores en la recopilación de datos, problemas de integridad de los datos o simplemente la falta de disponibilidad de información en ciertos campos.  
Es crucial detectar y comprender la distribución de los valores nulos o vacíos, ya que pueden afectar la validez y la eficacia de cualquier análisis posterior. La presencia de valores nulos puede distorsionar los resultados y conducir a interpretaciones incorrectas.  
Sin embargo, tomar decisiones sobre cómo manejar los valores nulos requiere un entendimiento profundo del negocio y del contexto específico de los datos. Dependiendo de la situación, se pueden adoptar diferentes estrategias, como eliminar las filas con valores nulos, imputar valores faltantes con técnicas estadísticas o de dominio, o incluso considerar los valores nulos como una categoría separada si tienen significado en el contexto del problema.
--**Valores atípicos**: También conocidos como outliers, son observaciones que se desvían significativamente del resto del conjunto de datos. Pueden surgir debido a errores de medición, eventos raros o simplemente representar datos genuinamente inusuales.  
Detectar valores atípicos es importante porque pueden distorsionar la interpretación de los resultados y afectar la precisión y la validez de los modelos estadísticos. Los valores atípicos pueden influir en medidas de tendencia central y dispersión, sesgar estimaciones de parámetros y reducir la eficacia predictiva de los modelos.  
Al igual que con los valores nulos, decidir cómo manejar los valores atípicos también requiere un profundo conocimiento del dominio y del contexto de los datos. En algunos casos, los valores atípicos pueden ser eliminados si se consideran errores o anomalías. Sin embargo, en otros casos, pueden contener información valiosa sobre situaciones excepcionales o patrones inesperados en los datos, por lo que es importante analizarlos cuidadosamente antes de tomar decisiones sobre su tratamiento.

En resumen, si bien es fundamental identificar y comprender los nulos, vacíos y valores atípicos en un conjunto de datos, la toma de decisiones sobre cómo manejarlos debe basarse en un profundo conocimiento del negocio y del contexto específico de los datos. No existe una solución única para todos los casos, y es importante considerar cuidadosamente las implicaciones de cada decisión en el análisis y la interpretación de los datos.

```python
# Se obtiene los tipos de datos de cada columna en el DataFrame dt_ini y lo convierte en un DataFrame. Luego, reinicia los indices del DataFrame resultante
tipo = dt_ini.dtypes.reset_index()
# Renombra los nombres de columnas
tipo.columns = ['VARIABLES','TIPO']

# Se calcula la cantidad de valores nulos en cada columna del DataFrame dt_ini y lo convierte en un DataFrame. Luego, eliminan los indices del DataFrame resultante
null = dt_ini.isnull().sum().reset_index()
# Renombra los nombres de columnas
null.columns = ['VARIABLES','NULOS']
# Se calcula el porcentaje de registros nulos sobre el total de registros 
null['%NULOS'] = null['NULOS']/dt_ini.shape[0]

# Se calcula la cantidad de valores vacios en cada columna del DataFrame dt_ini y lo convierte en un DataFrame. Luego, eliminan los indices del DataFrame resultante
vacio = (dt_ini=='').sum().reset_index()
# Renombra los nombres de columnas
vacio.columns = ['VARIABLES','VACIOS']
# Se calcula el porcentaje de registros vacios sobre el total de registros 
vacio['%VACIOS'] = vacio['VACIOS']/dt_ini.shape[0]

#### Estadisticas descriptivas a las variables numericas
# Se crea un DataFrame para guardar la informacion
numeric_info =pd.DataFrame()
# Itera sobre las columnas del DataFrame dt_ini que son de tipo numero o fecha, es decir, las variablesnumericas y fecha
for column in dt_ini.select_dtypes(include=['number','datetime']).columns:
    # Se verifica si es de tipo fecha/hora
    if dt_ini[column].dtype == 'datetime64[ns]':
        # se asigna None a la variable varianza ya que no se puede calcular la varianza sobre una fecha
        varianza = None
    # Si no es fecha se calcula la varianza de la columna y se asigna a varianza
    else:
        # calculo de varianza a variables numericas
        varianza = dt_ini[column].var() 
    # Se crea un DataFrame temporal que contiene las estadisticas descriptivas como el promedio, la mediana, el minimo, el maximo, la desviacion estandar y la varianza (si corresponde) para las columnas especificadas    
    #Este DataFrame temporal se concatena con el DataFrame numeric_info, y se repite este proceso para todas las columnas numéricas y de fecha/hora.
    numeric_info =pd.concat([numeric_info,pd.DataFrame({
        'VARIABLES':[column],
        'PROMEDIO':dt_ini[column].mean(), # Calculo del promedio
        'MEDIANA':dt_ini[column].median(), # Calculo de la mediana
        'MINIMO':dt_ini[column].min(), # Calculo del minimo
        'MAXIMO':dt_ini[column].max(), # Calculo del maximo
        'DESVIACION':dt_ini[column].std(), # Calculo de la desviacion estandar
        'VARIANZA': varianza
    })],ignore_index=True)

#### Estadisticas descriptivas a las variables categoricas
# Crear un DataFrame vacío para almacenar los resultados
cate_info = pd.DataFrame(columns=['VARIABLES', 'NUMERO_CATEGORIAS', 'NOMBRE_CATEGORIAS', 'CONTEO', '%CONTEO'])
# Definir el límite para el top n de categorías
top_n = 5
# Se itera sobre cada columna categorica
for columna in dt_ini.select_dtypes(include=['object', 'bool']).columns:
    # Se cuenta el numero de categorias unicas
    num_cate = dt_ini[columna].nunique()
    # Si hay mas de top_n categorías, obtener el top n
    if num_cate > top_n:
        # Se realiza el conteo de registros por categoria y se muestran los 5 mas altos
        conteo_cate = dt_ini[columna].value_counts().head(top_n)
    # Si no hay mas de top_n de categorias se muestran todas las categorias
    else:
        # Se realiza el conteo de registros por categoria
        conteo_cate = dt_ini[columna].value_counts()
    # Concatenar el DataFrame temporal al DataFrame principal
    cate_info = pd.concat([cate_info, pd.DataFrame({'VARIABLES': columna, 
                            'NUMERO_CATEGORIAS': num_cate, 
                            '%NUMERO_CATEGORIAS': ((num_cate / len(dt_ini)) * 100), # porcentaje de categorias sobre el total de registros                        
                            'NOMBRE_CATEGORIAS': conteo_cate.index, # nombre delas categorias
                            'CONTEO': conteo_cate.values,  # cantidad de registros por categoria
                            '%CONTEO': ((conteo_cate / len(dt_ini)) * 100).values # porcentaje de la categoria sobre el total de registros
      })], ignore_index=True)

# Se realiza una combinacion izquierda (left join) de los DataFrames tipo y null basada en la columna 'VARIABLES'. Es decir, combina los resultados de los tipos de variables y la cantidad de valores nulos
resul = pd.merge(tipo, null, on='VARIABLES', how='left') 
#Realiza una combinacion izquierda (left join) de los DataFrames resul y vacio basada en la columna 'VARIABLES'. Es decir, combina los resultados de resul con la cantidad de valores vacios
resul = pd.merge(resul, vacio, on='VARIABLES', how='left') 
# Realiza una combinacion izquierda (left join) de los DataFrames resul y los descriptivos de los datos numericos basada en la columna 'VARIABLES'
resul = pd.merge(resul, numeric_info, on='VARIABLES', how='left') 
# Realiza una combinacion izquierda (left join) de los DataFrames resul y cate_info basada en la columna 'VARIABLES'. Es decir, combina los resultados de resul con las estadisticas descriptivas categoricas.
resul = pd.merge(resul, cate_info, on='VARIABLES', how='left') 

# Se guarda el DataFrame resul en un archivo Excel llamado 'descriptivos.xlsx' sin incluir los índices de fila.
resul.to_excel('descriptivos.xlsx',index=False)

# Se configura el filtro de advertencias para ignorarlas.
warnings.filterwarnings("ignore")
```

### Cálculo de nuevas variables

El cálculo de nuevas variables en un proyecto de ciencia de datos se realiza con el fin de enriquecer la información disponible y proporcionar características adicionales que puedan ser útiles para el análisis o la construcción de modelos. Algunas razones para realizar este cálculo incluyen:

- Captura de relaciones complejas: Las nuevas variables pueden capturar relaciones más complejas entre las características existentes, lo que permite al modelo aprender patrones más sutiles y precisos en los datos.

- Reducción de la dimensionalidad: A veces, la creación de nuevas variables puede ayudar a reducir la dimensionalidad de los datos al resumir información redundante o eliminar características menos importantes.

- Adaptación a los requisitos del modelo: Algunos modelos pueden beneficiarse de ciertas características específicas. La creación de nuevas variables puede adaptar los datos para cumplir con los requisitos de un modelo particular.

En cuanto a la conversión de una variable categórica con muchas categorías y vacíos a una variable binaria (sí o no), este proceso se conoce como binarización o codificación binaria. Esto se hace por varias razones:

- Simplificación de la información: La binarización reduce la complejidad al convertir una variable categórica con múltiples categorías en una variable binaria simple, lo que facilita su interpretación y análisis.

- Manejo de vacíos: Al convertir la variable en una variable binaria, se puede identificar fácilmente si una observación tiene o no información en esa variable, lo que puede ser útil para manejar los valores faltantes de manera efectiva.

- Preparación para modelos de aprendizaje automático: Muchos algoritmos de aprendizaje automático requieren que las variables de entrada sean numéricas. La binarización facilita el uso de estas variables categóricas en modelos de clasificación o segmentación.

En un modelo de segmentación, la conversión de variables categóricas a variables binarias puede ser especialmente útil para identificar patrones distintivos en los datos y segmentar de manera efectiva el conjunto de observaciones en grupos homogéneos. Al representar las características relevantes de manera binaria, se facilita la identificación de las características que tienen un mayor impacto en la segmentación y se simplifica el proceso de interpretación de los resultados del modelo.

| combinations | itertools | Genera todas las combinaciones únicas de longitud 2 de las columnas de variables categóricas para realizar la prueba de chi-cuadrado. |
| append | Python Built-in | Agrega los resultados de la prueba de chi-cuadrado (estadístico, valor p, rechazo de H0) a la lista de resultados. |

```python
# Se crea una lista para almacenar los resultados
resul_chi = []
# Se selecciona solo las columnas con variables categoricas
col_cat = dt_pre1.select_dtypes(include=['object'])
# Se excluye el id del producto
col_cat = col_cat.drop(columns=['id'])
# Se genera todas las combinaciones unicas de variables categoricas
combinaciones = list(com(col_cat.columns, 2))
# Se itera todas las combinaciones
for col1, col2 in combinaciones:
    # Se calcula la tabla de frecuencias cruzadas
    tabla = pd.crosstab(col_cat[col1], col_cat[col2])
    # Se realiza la prueba de chi-cuadrado para cada combinacion
    estadistico, p_value, _, _ = chi(tabla)
    # Se condiciona donde si el valor p es menor a 0.05 entonces 'SI' rechaza H0 de lo contrario 'NO' rechaza H0
    rechaza_H0 = "SI" if p_value < 0.05 else "NO"
    # Se agregan todos los resultados a un mismo DataFrame
    resul_chi.append({
        "Variable_1": col1,
        "Variable_2": col2,
        "Estadistico": estadistico,
        "Valor_P": p_value,
        "Rechaza_H0": rechaza_H0
    })

# Convertir la lista de resultados a un DataFrame
resul_df = pd.DataFrame(resul_chi)
# Mostrar la tabla de resultados
resul_df
```

### Clustering
Los modelos de segmentación, también conocidos como modelos de clustering, son técnicas de aprendizaje no supervisado que se utilizan para agrupar datos similares en conjuntos distintos, llamados "clusters". Estos modelos son ampliamente utilizados en diversas aplicaciones, desde la segmentación de clientes hasta la organización de datos no etiquetados.

#### Variables categóricas en dicotomicas

La conversión de datos categóricos a dicotómicos es necesaria en algunos casos para aplicar modelos de segmentación debido a que muchos algoritmos de clustering están diseñados para trabajar con variables numéricas. Para modelos específicos como K-Prototypes con Hamming y Self-Organizing Maps (SOM), la conversión de datos categóricos a dicotómicos es esencial por las siguientes razones:

- K-Prototypes con Hamming: En K-Prototypes, que es una variante del algoritmo K-Means diseñada para manejar datos mixtos (numéricos y categóricos), la distancia entre puntos de datos categóricos se mide utilizando la distancia de Hamming. Esta distancia de Hamming es una métrica que compara la discrepancia entre dos cadenas de símbolos de igual longitud. Para calcularla adecuadamente, las variables categóricas se convierten en variables binarias (dicotómicas), donde cada categoría se representa como un valor binario (0 o 1). De esta manera, se puede calcular la distancia de Hamming entre dos puntos de datos categóricos de manera efectiva.
Self-Organizing Maps (SOM): En el caso de SOM, que es un algoritmo de aprendizaje no supervisado utilizado para visualizar y agrupar datos de alta dimensión, las entradas también deben ser numéricas para calcular las distancias entre nodos y actualizar los pesos de manera adecuada durante el entrenamiento. Al convertir las variables categóricas a dicotómicas, se pueden representar como vectores binarios, lo que permite calcular las distancias entre nodos en el espacio de características de manera eficiente y realizar las actualizaciones necesarias durante el entrenamiento del SOM.

Para utilizar modelos como K-Prototypes con Hamming y Self-Organizing Maps (SOM), es necesario convertir los datos categóricos a dicotómicos para poder calcular distancias y realizar actualizaciones de pesos de manera efectiva durante el proceso de clustering. Esto asegura que estos modelos puedan manejar datos mixtos y producir resultados significativos y precisos.

#### K-Prototypes con Hamming
El algoritmo K-Prototypes con Hamming es una extensión del algoritmo K-Prototypes diseñado específicamente para manejar conjuntos de datos mixtos que contienen tanto variables numéricas como categóricas. Mientras que el K-Prototypes estándar utiliza una medida de distancia apropiada para variables numéricas (por ejemplo, la distancia euclidiana) y otra medida para variables categóricas (por ejemplo, la distancia de Gower), el K-Prototypes con Hamming utiliza la distancia de Hamming para las variables categóricas.

La distancia de Hamming mide la diferencia entre dos cadenas de igual longitud, contando el número de posiciones en las que los símbolos correspondientes son diferentes. En el contexto del clustering, esta distancia se utiliza para comparar las categorías de las variables categóricas y calcular la distancia entre diferentes puntos de datos.

Así que, el K-Prototypes con Hamming combina la distancia de Hamming para variables categóricas y una medida adecuada para variables numéricas para agrupar eficazmente datos mixtos en grupos homogéneos. Es útil en situaciones donde se necesitan análisis de clustering en conjuntos de datos que incluyen tanto características categóricas como numéricas.

#### Método del Codo para la selección de k
El método del codo es una técnica utilizada para seleccionar el número óptimo de clusters (k) en un algoritmo de clustering. Se basa en calcular la suma de los cuadrados de las distancias de cada punto de datos al centroide más cercano dentro del cluster al que pertenece. El objetivo es identificar el punto en el gráfico donde la tasa de disminución de la suma de los cuadrados de las distancias disminuye significativamente, formando un "codo".

Para aplicar el método del codo:
- Se ajusta el algoritmo de clustering (por ejemplo, KMeans) con diferentes valores de k.
- Se calcula la suma de los cuadrados de las distancias para cada valor de k.
- Se traza un gráfico de la suma de los cuadrados de las distancias en función de k.
- Se identifica el punto en el gráfico donde se produce una "curva de codo" o un cambio significativo en la pendiente.
- El valor de k en el punto del codo se selecciona como el número óptimo de clusters.
- Este método se considera útil porque permite identificar un valor de k que equilibra la complejidad del modelo (número de clusters) con la capacidad de explicar la varianza en los datos.

El método de la silueta también se utiliza para seleccionar el número óptimo de clusters. Calcula la cohesión dentro de los clusters y la separación entre los clusters. El valor de la silueta varía entre -1 y 1, donde un valor más cercano a 1 indica que el punto está bien agrupado dentro de su cluster y mal clasificado en los clusters vecinos. El método de la silueta selecciona el valor de k que maximiza el valor medio de la silueta sobre todos los puntos de datos. Sin embargo, a veces puede ser más subjetivo y menos claro determinar el número óptimo de clusters utilizando este método, especialmente cuando los clusters tienen formas irregulares o superposiciones. Por lo tanto, el método del codo es preferido en muchos casos debido a su simplicidad y facilidad de interpretación.

### Self-Organizing Maps (SOM)

Los Mapas Auto-Organizados (Self-Organizing Maps o SOM, por sus siglas en inglés) son una técnica de aprendizaje no supervisado utilizada en el campo del aprendizaje automático y la minería de datos. Fueron desarrollados por Teuvo Kohonen en la década de 1980.

Un SOM es un tipo de red neuronal artificial que mapea relaciones complejas y no lineales entre los datos de entrada en una estructura bidimensional (o tridimensional) de nodos o neuronas, conocida como mapa SOM. Cada neurona en el mapa representa un vector de pesos que se ajusta durante el proceso de entrenamiento para adaptarse a las características de los datos de entrada.

El objetivo principal de un SOM es agrupar datos similares en regiones cercanas del mapa, lo que permite visualizar y entender la estructura subyacente de los datos. Para lograr esto, durante el entrenamiento, los pesos de las neuronas se ajustan de manera que las neuronas cercanas en el mapa tengan vectores de pesos similares, mientras que las neuronas distantes tengan vectores de pesos diferentes. Esto se logra mediante un proceso iterativo de competencia y cooperación entre las neuronas, donde las neuronas compiten por activarse en respuesta a los datos de entrada y luego cooperan para adaptarse a los patrones de entrada.

Los SOM son útiles para la visualización de datos de alta dimensión, el descubrimiento de patrones en conjuntos de datos grandes y la reducción de la dimensionalidad para facilitar el análisis y la interpretación de datos complejos. También se utilizan en diversas aplicaciones como la clasificación de imágenes, la segmentación de clientes, la detección de fraudes y la comprensión de datos biológicos y ambientales, entre otros.

Las neuronas en un mapa auto-organizado (SOM) representan unidades de procesamiento que se organizan en una cuadrícula bidimensional. Estas neuronas compiten entre sí para aprender y representar la estructura de los datos de entrada de manera topológica. La razón por la que se definen las variables x_dim y y_dim, que representan las dimensiones del SOM, es para especificar el tamaño de esta cuadrícula de neuronas. En el caso específico del código proporcionado, se establece un SOM de 10x10 neuronas (x_dim = 10 y y_dim = 10). Esto significa que el SOM tendrá 10 neuronas en la dimensión horizontal (eje x) y 10 neuronas en la dimensión vertical (eje y). En otras palabras, el SOM consistirá en una cuadrícula de 10x10 neuronas, lo que permite capturar una representación bidimensional de la estructura de los datos de entrada. Este tamaño de SOM se elige según la complejidad y la dimensionalidad de los datos, así como los requisitos específicos del problema que se está abordando. Un SOM más grande puede capturar una representación más detallada de los datos, pero también puede requerir más recursos computacionales y datos de entrenamiento.

```python
# Se definen las variables x_dim y y_dim, que representan las dimensiones del mapa auto-organizado. En este caso, se establece un SOM de 10x10, lo que significa que habrá una cuadrícula de 10x10 neuronas en el SOM
x_dim = 10
y_dim = 10

# Se crea un nuevo objeto SOM. input_len representa la longitud de las caracteristicas de entrada del conjunto de datos preprocesado
som = MiniSom(x=x_dim, y=y_dim, input_len=dt_preprocesado.shape[1])
# Se entrena el SOM con los datos preprocesados del conjunto de entrenamiento. El segundo argumento (100) especifica el nUmero de iteraciones de entrenamiento
datos_numpy = dt_preprocesado.values.astype(float)
som.train(datos_numpy, 100)
# Se obtienen los clusters asignados por el SOM. Se utilizan los datos preprocesados para asignar cada muestra al nodo ganador mas cercano en el SOM
cluster_som = som.win_map(datos_numpy)
# Se obtienen las caracteristicas mas importantes para cada neurona
componentes = som.get_weights()
```

```python
# Visualizar los componentes de las neuronas
# Se crea una figura de Matplotlib con una dimensión específica de 10x6
plt.figure(figsize=(10, 6))
# Se visualizan los componentes de las neuronas como un mapa de calor, donde se especifican los componentes, el color del mapa, la anotacion para mostrar los valores en cada celda,el formato de los valores, y se habilita la barra de color
sns.heatmap(componentes[:, :, 0], cmap='viridis', annot=True, fmt='.2f', cbar=True)
# Se establece el titulo
plt.title('Visualización de Componentes de Neurona')
# Se nombra el eje x
plt.xlabel('Dimensiones de las Características')
# Se nombra el eje y
plt.ylabel('Neuronas')
# Se muestra la figura
plt.show()
```

Se utiliza un mapa de calor (heatmap) para representar estos componentes. Cada celda en el mapa de calor corresponde a una neurona en el SOM, y el color de cada celda indica el valor del componente para esa neurona.

Los valores más altos indican una mayor importancia de esa característica para la neurona correspondiente. Esto significa que las características con valores más altos tienen un impacto significativo en la activación de esa neurona.
Los valores más bajos indican una menor importancia de esa característica para la neurona correspondiente. Esto significa que las características con valores más bajos tienen menos influencia en la activación de esa neurona.
La interpretación de la importancia de los valores bajos y altos depende del contexto del problema y de la naturaleza de las características. En general, las características con valores altos pueden considerarse como características distintivas o discriminativas que ayudan a definir la identidad de una neurona específica en el SOM. Por otro lado, las características con valores bajos pueden ser menos relevantes o informativas para la agrupación de datos en esa neurona en particular.

Esta visualización ayuda a entender qué características son más importantes para cada neurona en el SOM, lo que puede ser útil para interpretar cómo se agrupan y organizan los datos en el espacio de características definido por el SOM.
